### Question 4: Data Cleaning

1. List the samples in the **cars** that have missing values in the attribute `Max.Price`.
2. Fill in the missing values in the attribute `Max.Price` with the attribute's average value.
3. Detect any duplicate records. If duplicates are found, remove them, ensuring only one instance of each duplicate record is retained.
4. Remove the samples in the **cars** that have missing values in the attribute `Horsepower`.

```python
# your answer
missing_max_price = cars[cars['Max.Price'].isna()]
print(missing_max_price)
```

```python
average_max_price = cars['Max.Price'].mean()
cars['Max.Price'].fillna(average_max_price, inplace=True)
```

```python
duplicates = cars.duplicated()
print(f"Number of duplicate records: {duplicates.sum()}")
cars.drop_duplicates(inplace=True)
```

```python
cars = cars.dropna(subset=['Horsepower'])
```

### Question 5: Data Normalization

Normalize the attribute `Max.Price` using following methods:

1. Min-max normalization.
2. Z-score normalization.

```python
# your answer
max_price_min = cars['Max.Price'].min()
max_price_max = cars['Max.Price'].max()
cars['Max.Price_MinMax'] = (cars['Max.Price'] - max_price_min) / (max_price_max - max_price_min)
```

```python
max_price_mean = cars['Max.Price'].mean()
max_price_std = cars['Max.Price'].std()
cars['Max.Price_ZScore'] = (cars['Max.Price'] - max_price_mean) / max_price_std

cars[['Max.Price', 'Max.Price_MinMax', 'Max.Price_ZScore']].head()
```

### Question 6: Data Discretization

Transform the `Max.Price` attribute into a discrete categorical feature with three categories using the following discretization methods:

1. **Equal-width binning:** Divide the range of `Max.Price` into intervals of equal size.
2. **Equal-depth binning:** Distribute the `Max.Price` values into bins so that each bin contains roughly the same number of records.

```python
# your answer
num_bins = 3
cars['Max.Price_EqualWidth'] = pd.cut(cars['Max.Price'], bins=num_bins)
```

```python
cars['Max.Price_EqualDepth'] = pd.qcut(cars['Max.Price'], q=num_bins)
```

```python
cars[['Max.Price', 'Max.Price_EqualWidth', 'Max.Price_EqualDepth']].head()
```

### Question 7: Data Visualization

Visualize the `Max.Price` attribute with the following plots:
- Create Box Plots to examine the distributions of `Max.Price` and `Horsepower`.
- Generate Histograms (using 50 bins) to observe the frequency distributions of `Max.Price` and `Horsepower`.
- Construct a Scatter Plot to explore the relationship between `Max.Price` and `Horsepower`.

```python
import matplotlib.pyplot as plt
```

```python
# your answer
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
cars.boxplot(column='Max.Price')
plt.title('Box Plot of Max.Price')

plt.subplot(1, 2, 2)
cars.boxplot(column='Horsepower')
plt.title('Box Plot of Horsepower')

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
cars['Max.Price'].hist(bins=50)
plt.title('Histogram of Max.Price')

plt.subplot(1, 2, 2)
cars['Horsepower'].hist(bins=50)
plt.title('Histogram of Horsepower')

plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(cars['Horsepower'], cars['Max.Price'])
plt.title('Scatter Plot of Max.Price vs Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Max.Price')
plt.show()
```
